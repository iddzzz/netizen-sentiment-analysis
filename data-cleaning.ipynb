## Concatenating Data

```python
import pandas as pd
import os
from langdetect import detect
pd.set_option("display.max_colwidth", 140) # to set the column width of the table output
```

```python
tw1 = pd.read_csv("./data/tweet-01.csv")
tw2 = pd.read_csv("./data/tweet-02.csv")
tw3 = pd.read_csv("./data/tweet-03.csv")
tw4 = pd.read_csv("./data/tweet-04.csv")
tw5 = pd.read_csv("./data/tweet-05.csv")
tw6 = pd.read_csv("./data/tweet-06.csv")
tw7 = pd.read_csv("./data/tweet-07.csv")
tw8 = pd.read_csv("./data/tweet-08.csv")

tw9_1 = pd.read_csv("./data/tweet-09-05-01.csv")
tw9_2 = pd.read_csv("./data/tweet-09-07-04.csv")
tw9_3 = pd.read_csv("./data/tweet-09-10-06.csv")
tw9_4 = pd.read_csv("./data/tweet-09-30-09.csv")

tw9 = pd.concat([tw9_1, tw9_2, tw9_3, tw9_4])

tw10 = pd.read_csv("./data/tweet-10.csv")
tw11 = pd.read_csv("./data/tweet-11.csv")
tw12 = pd.read_csv("./data/tweet-12.csv")

tw10_11 = pd.concat([tw10, tw11])
tw10_11 = tw10_11.drop_duplicates()
# tw10_11[tw10_11.duplicated(keep=False)].sort_values('id')
tw10_12 = pd.concat([tw10_11, tw12])
tw10_12 = tw10_12.drop_duplicates()
# tw10_12[tw10_12.duplicated(keep=False)].sort_values('id')
tw9 = tw9.drop_duplicates()
# tw9[tw9.duplicated(keep=False)].sort_values('id')
tw9_12 = pd.concat([tw9, tw10_12])
tw9_12 = tw9_12.drop_duplicates()
tw9_12[tw9_12.duplicated(keep=False)].sort_values('id')
```

```python
df = pd.concat([tw1, tw2, tw3, tw4, tw5, tw6, tw7, tw8, tw9_12])
df = df.drop_duplicates()
df[df.duplicated(keep=False)].sort_values('id')
```

```python
print(df['date'][df.date == 'Aug 2, 2021'])
def convert_date(text: str):
    month_dict = {
        'Jan': '01', 'Feb': '02', 'Mar': '03', 'Apr': '04',
        'May': '05', 'Jun': '06', 'Jul': '07', 'Aug': '08',
        'Sep': '09', 'Oct': '10', 'Nov': '11', 'Dec': '12'
    }
    bulan = text.split(' ')[0]
    bulan = month_dict[bulan]
    tahun = text.split(' ')[-1]
    tanggal = text.split(',')[0].split(' ')[-1]
    tanggal = tanggal if len(tanggal) == 2 else ('0' + tanggal)
    return f'{tahun}-{bulan}-{tanggal}'
```

```python
df.date = df.date.apply(convert_date)
df.date = pd.to_datetime(df.date)
df.info()
```

### Extract mentions

```python
def extract_mentions(text):
    mentions = []
    text = text.split('\n')
    for item in text:
        for i in item.split():
            if i[0] == '@':
                mentions.append(i)
    return ' '.join(mentions)
```

```python
df['mentions'] = df.content.apply(extract_mentions)
df[['content', 'mentions']][df.mentions != '']
```

## content_clean

```python
content_clean = []
# Removing twitter mentions and links
for _, row in df.iterrows():
    content = row['content']
    mentions = row['mentions'].split()
    for mention in mentions:
        content = content.replace(mention, '')
    content = content.replace('\n', ' ')
    content = content.split()
    for i, item in enumerate(content):
        if 'http' in item:
            content.pop(i)
    content = ' '.join(content)
    content_clean.append(content)
df['content_clean'] = content_clean
```
