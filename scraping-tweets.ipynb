# Data Scraping from Twitter

```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

## December

```python
data_path = os.path.join(os.path.realpath(os.getcwd()), 'data')
dec = pd.read_csv(os.path.join(data_path, 'tweet12.csv'))

# Drop the data from 1 Jan 2022
dec = dec.drop(0, axis=0)

# There are several abnormal tweet structure
def tweet_normalkah(tweet):
    try:
        return tweet.split('\n')[3].split(' ')[2] == '2021'
    except Exception as why:
        return False

# A column indicating whether the tweet normal
dec['tw_normal'] = dec.Tweet.apply(lambda x: 1 if tweet_normalkah(x) else 0)

# Retreive the date
dec['tanggal'] = dec.Tweet.apply(lambda x: x.split('\n')[3] if tweet_normalkah(x) else x.split('\n')[2])
# data[data['tw_normal'] == 0]

dec = dec.drop('Unnamed: 0', axis=1)
dec
```

```python
sns.barplot(x=dec.tw_normal.value_counts().index, y=dec.tw_normal.value_counts())
plt.xlabel("type")
plt.ylabel("counts")
plt.title("Comparison")
plt.show()
```

## Concatenate All

```python
df = pd.concat([jan, feb, mar, apr, may, jun, jul, aug, sep, octo, nov, dec])
df.reset_index(inplace=True)
df.drop('index', axis=1, inplace=True)
df
```

```python
def apakah_normal_juga(tanggal):
    try:
        return tanggal.split(' ')[2] == '2021'
    except Exception as why:
        return False

df['tw_normal_2'] = df.tanggal.apply(lambda x: 1 if apakah_normal_juga(x) else 0)
df[df.tw_normal_2 == 0].tail(10)
```

The tw_normal_2 indicates advertisements

```python
df = df[df.tw_normal_2 == 1]
df.drop('tw_normal_2', axis=1, inplace=True)
df
```

```python
df_abnormal = df[df.tw_normal == 0].copy()
df_normal = df[df.tw_normal == 1].copy()
print(df_abnormal.shape, df_normal.shape)
```

```python
df_abnormal.Tweet = df_abnormal.Tweet.apply(lambda x: '\n'.join([x.split('\n')[0].replace('@', ''), x]))
```

```python
df = pd.concat([df_normal, df_abnormal])
df.reset_index(inplace=True)
df.drop('index', axis=1, inplace=True)
```

```python
df['name'] = df.Tweet.apply(lambda x: x.split('\n')[0].strip())
df['username'] = df.Tweet.apply(lambda x: x.split('\n')[1].strip())
df['at'] = df.username.apply(lambda x: 1 if x[0] == '@' else 0)
```

```python
analyse = df[df['at'] == 0].copy()
analyse.sample().Tweet.values
```

```python
# print(df.sample(5)['Tweet'].values)

# Apakah tweet adalah reply
def is_reply(text):
    return text.split('\n')[4].split(' ')[0]

df['reply'] = df.Tweet.apply(lambda x: 1 if is_reply(x) == 'Replying' else 0)
df[df['reply'] == 1].sample(5).values
```
